```python
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from matplotlib import pyplot as plt
```

```python
# Função para exibir uma imagem
def print_img(img):
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.axis('off')
    plt.show()
    
```

```python
img = cv2.imread('lisa.jpg', cv2.IMREAD_GRAYSCALE)
print(img)          # Mostra a matriz da imagem
print(type(img))    # Mostra o tipo da variável img
print(len(img))     # Mostra o número de linhas da imagem
print(len(img[0]))  # Mostra o número de colunas da imagem
print(img.shape)    # Mostra as dimensões da imagem
print_img(img)      # Exibe a imagem
```

```python
# Redimensionar a imagem
img_r = cv2.resize(img, (64, 64))   # Redimensiona para 64x64 pixels
print_img(img_r)                    # Exibe a imagem redimensionada
```

```python
# Normalizar a imagem
img_n = img_r.astype('float32') / 255.0 # Normaliza os valores para o intervalo [0, 1]
print_img(img_n)                        # Exibe a imagem normalizada
```

```python
import scipy.signal

# Filtro para detecção de bordas horizontal
kernel = np.array([
    [-1, -1, -1],
    [ 0,  0,  0],
    [ 1,  1,  1]
])

# Execultar o filtro na imagem normalizada com o kernel definido
filtro = scipy.signal.convolve2d(img_n, kernel, mode='same')

# Normalizar no resultado filtrado para visualização
filtro_normalizado = (filtro - filtro.min()) / (filtro.max() - filtro.min())

plt.imshow(filtro_normalizado, cmap='gray')
plt.title('Saída da Convolução (Filtro de Bordas Horizontal)')
plt.axis('off')
plt.show()
```

```python
import scipy.signal

# Filtro para detecção de bordas vertical
kernel = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1]
])

# Execultar o filtro na imagem normalizada com o kernel definido
filtro = scipy.signal.convolve2d(img_n, kernel, mode='same')

# Normalizar no resultado filtrado para visualização
filtro_normalizado = (filtro - filtro.min()) / (filtro.max() - filtro.min())

plt.imshow(filtro_normalizado, cmap='gray')
plt.title('Saída da Convolução (Filtro de Bordas Horizontal)')
plt.axis('off')
plt.show()
```
